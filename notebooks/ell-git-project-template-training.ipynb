# Install Training Code & Dependencies

```python
! pip freeze > reqs-existing.txt
! git clone https://github.com/mrcabbage972/kaggle-ell-public.git
! pip config set global.disable-pip-version-check true
! cd kaggle-ell-public && cd src && pip install -q .
! cd kaggle-ell-public && git rev-parse --short HEAD > ../git_version.txt
#! wandb login [YOUR WANDB API KEY HERE]
```

# Run Training

```python
! run_solution \
git_hash="$(<git_version.txt)" \
env=kaggle \
solution=transformer_finetune \
solution.do_train=true \
solution.args.train.epochs=3 \
solution.args.train.trn_fold=[0,1,2,3] \
solution.args.model.backbone=microsoft/deberta-v3-base
```

# Create Python Wheels for Inference
This purpose of the below code is to create a package that can be installed in the inference notebook.

```python
! pip freeze > reqs-package.txt
! grep -v -x -f reqs-existing.txt reqs-package.txt > reqs-final.txt
! cat reqs-final.txt
! cd kaggle-ell-public/src && pip wheel -q . -r /kaggle/working/reqs-final.txt -w /kaggle/working/wheels --no-deps
! rm -r kaggle-ell-public
```
